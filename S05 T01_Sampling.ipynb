Un cop tenim el DataFrame net, podem procedir al sampling demanat:

```python
#Generem mètodes per a obtenir les mostres de forma aleatòria

def aleatori_sampling(population,samples):#population and sample són integers!
    sampling_list=[]
    i=0
    while i <samples:
        position=int(random.random()*population)
        if position not in sampling_list:
            sampling_list.append(position)
            i+=1
    return sampling_list    
```

```python
sampling_list =aleatori_sampling(df_scoring.shape[0],100)
sample_df=df_scoring.iloc[sampling_list]
sample_df
```

Després veig que la mostra aleatoria es pot obtenir directament amb el mètode 'sample' de Pandas:

```python
sample_df2 = df_scoring.sample(100)
sample_df2
```

```python
#Generem mètode per a obtenir les mostres de forma sistemàtica
def sistematic_sampling(population,samples):
    sampling_list=[]
    z=population/samples
    position=int(random.random()*z)
    sampling_list.append(position)
    for i in range(1,samples):
        sampling_list.append(int(i*z)+position)
    return sampling_list
```

```python
sampling_list2 =sistematic_sampling(df_scoring.shape[0],100)
sample_df3=df_scoring.iloc[sampling_list2]
sample_df3
```

Procedim a a fer un **mostrejat amb SMOTE**:

```python
from collections import Counter
from imblearn.over_sampling import SMOTENC
```

```python
sm = SMOTENC(random_state=42, categorical_features=[0])
X=df_scoring[['lgID','G','A']].sample(10000) #estudiem relació gols (G) i assistencies (A) per equip.
y=list(X['lgID'])
```

```python
X.shape #tamany de la mostra: conté 10000 registres diferents amb 3 atributs per registre (1 categòric).
```

```python
Counter(y)
```

Clarament hi ha un desbalanç entre les lligues. Mitjançant Smotenc generem dates per ígualar el número de registres per cada equip, generarà valors en funció dels valors veinals. En aquest cas segurament no té massa lògica...

```python
X_res, y_res = sm.fit_resample(X, y)
```

Continua amb el conjunt de dades de tema esportiu i genera una mostra utilitzant el mètode Reservoir sampling.

Video ilustrador :
    https://youtu.be/A1iwzSew5QY

Teoria y codi python:
    https://www.geeksforgeeks.org/reservoir-sampling/

```python
def reservoir_sampling(population, samples): #Entrada= # d'elements a la població i # d'elements a la mostra       
        cont=0;
        reservoir = [0]*samples #generem llista amb 0 per a seguidament canviar-los
        for i in range(samples):
            reservoir[i] = i
    
        while(cont < population):
            # Agafa un numero aleatori d'un rang que va creixent de 0 a population-1
            j = random.randrange(cont+1)
             
            # Si el número aleatori j és menor que el número d'elements al sampling (=samples), llavors canviarà la posició obtinguda j 
            # amb el valor de cont. Si j és major que samples, llavors no canvia la llista reservoir i suma un a cont 
            
            if(j < samples):
                reservoir[j] = cont
            cont+=1;
         
        return reservoir #retorna l'index del Reservoir sampling
    
```

```python
# Aplicat a la base de dades de Hockey (tamany del sampling=100)
 
reservoir_Sample_df=df_scoring.iloc[reservoir_sampling(df_scoring.shape[0],100)]
reservoir_Sample_df
```

```python
reservoir_Sample_df.info()
```

Es mostreja de forma més ràpida sense revisar si els valors ja estaven a la llista degut a la metodologia de selecció dels registres. Aquesta metodologia també assegura que la provabilitat d'escollir un registre o un altre és igual per a tots els registres.
